# Load Data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#get single train_set
raw_data = pd.read_csv("../Data/RUL_Train_001.csv")

raw_data
```

# PCA

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = ['ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 
            'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 
            'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 
            'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 
            'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 
            'SensorMeasure13', 'SensorMeasure14', 'SensorMeasure15', 
            'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 
            'SensorMeasure19', 'SensorMeasure20', 'SensorMeasure21']

# Separating out the features
x = raw_data.iloc[:, 0:-1].values

# Separating out the target
y = raw_data.loc[:,['RUL']].values

# Fit on raw_data
scaler.fit(x)

# Apply transform to raw_data
x = scaler.transform(x)
```

```python
from sklearn.decomposition import PCA

# Make an instance of the Model
pca = PCA(.95) #95% of the variance (information) is retained

pca.fit(x)

x = pca.transform(x)
```

```python
fir = plt.figure(figsize=(8,5))
sing_vals = np.arange(len(pca.components_)) + 1
plt.plot(sing_vals, pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.title('Scree Plot', fontsize = 20)
plt.xlabel('Principal Component', fontsize = 20)
plt.ylabel('Eigenvalue', fontsize = 20)
```

```python
principalDf = pd.DataFrame(data = x)
RULDF = pd.DataFrame(data = y)
```

```python
finalDf = pd.concat([principalDf, RULDF], axis = 1)
finalDf.columns = [*finalDf.columns[:-1], 'RUL']

finalDf
```
